```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df = pd.read_csv("Market Basket_Small dataset.csv")
display(df.head(15))
print(df.shape)
```

```python
df.info()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Convert DataFrame to a list of lists and convert values to strings
transactions = df.astype(str).values.tolist()

# Convert the transaction data to a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Display the association rules
print(rules)
```
